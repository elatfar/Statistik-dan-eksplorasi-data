```python
import random

sampel_mean = []
counter = 1
sepal_length = iris.data[:,0]

for i in range(10):
    y = random.sample(sepal_length.tolist(),20)
    avg = np.mean(y)
    print("boostrap "+str(counter)+" : "+str(avg))
    counter+=1
    sampel_mean.append(avg)

print("\n")
print(sampel_mean)    
print(np.mean(sampel_mean))
```

```python
def load_excel(path_excel,awal,potong):
    import pandas as pd
    
    data = pd.read_excel(path_excel)
    return data[awal,potong]
```

```python
load_excel("excel_1.xls",awal=1,potong=100)
load_excel("excel_1.xls",awal=101,potong=200)
```
